```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = 1
K = 99
S = 100
sigma = 0.2
r = 0.06
```

```python
def GBM_Euler(T, S, sigma, r, M, Z_all = None):
    S_all = []
    S_all.append(S)
    dt = T/M
    
    Zm = np.random.normal(size=M)
    for i in range(M-1):
        S_all.append(S_all[i] + r* S_all[i] * dt + sigma * S_all[i] * np.sqrt(dt) * Zm) 
    return S_all
```

```python
M=365
s_all = [GBM_Euler(T, S, sigma, r, M) for x in range(1000)]
```

```python
plt.plot(s_all[0])
```
